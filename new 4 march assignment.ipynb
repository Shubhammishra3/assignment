```python
import seaborn as sns
```

```python
tit = sns.load_dataset("titanic")
x = tit["age"]
y = tit["fare"]
```

```python
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatter(x = x , y = y , mode = "markers"))
fig.show()
```

```python
"""
Q5. What is Distplot? Using Plotly express, plot a distplot.


answer]The distplot figure factory displays a combination of statistical representations of numerical data, such as histogram,
kernel density estimation or normal curve, and rug plot.

The distplot can be composed of all or any combination of the following 3 components −

1] histogram
2]  curve: (a) kernel density estimation or (b) normal curve, and
3]  rug plot
```

```python
import numpy as np
```

```python
import plotly.figure_factory as ff
import numpy as np

x1 = np.random.randn(200) - 1
x2 = np.random.randn(200)
x3 = np.random.randn(200) + 1

hist_data = [x1, x2, x3]

group_labels = ['Group 1', 'Group 2', 'Group 3']
colors = ['#333F44', '#37AA9C', '#94F3E4']

# Create distplot with curve_type set to 'normal'
fig = ff.create_distplot(hist_data, group_labels, show_hist=True, colors=colors)

# Add title
fig.update_layout(title_text='Curve and Rug Plot')
fig.show()
```
